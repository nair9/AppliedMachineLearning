```python
# Simulated data is given as follows in Python:

import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)
plt.plot(x,y,'ro')
```

```python
# 1. Please plot the noisy data and the polynomial you found (in the same figure). You can use any
# value of m selected from 2, 3, 4, 5, 6.

# for m = 6

z = np.polyfit(x.flatten(),y.flatten(),6)
p = np.poly1d(z)
x2 = np.linspace(-15,15,50)

plt.plot(x.flatten(),y.flatten(),'ro',x2,p(x2),'b--')

plt.title('Plot of the polynomial')
plt.legend(['noisy data','polynomial'])
```

```python
# 2. Plot MSE versus order m, for m = 1, 2, 3, 4, 5, 6, 7, 8 respectively. Identify the best choice of m.

noise_scale = 100
number_of_samples = 50
x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
y = 5 * x + 15 * x**2 + 2 * x**3 + noise_scale*np.random.randn(number_of_samples, 1)

x = x.flatten()
y = y.flatten()
m = [1, 2, 3, 4, 5, 6, 7, 8]

z = {}
p = {}
mse = {}

for i in range(1,9):
    z[i] = np.polyfit(x,y,i)
    p[i] = np.poly1d(z[i])
    mse[i] = (sum((y-p[i](x))**2))/number_of_samples
    
plt.plot(m,mse.values(),'b--')
plt.title('Plot of MSE versus order m')
```

The best value of m from the plot above is :- m = 3

```python
# 3. Change variable noise_scale to 200, 300, 400, 600, 800, 1000 respectively, re-run the algorithm and plot the polynomials with the m found in 2). Discuss the impact of noise scale to the
# accuracy of the returned parameters. [You need to plot a figure like in 1) for EACH choice of noise_scale.]

noise_scale = [200, 300, 400, 600, 800, 1000]
number_of_samples = 50

x = 30*(np.random.rand(number_of_samples, 1) - 0.5)
x1 = x.flatten()
m = 3

for i in noise_scale:
    y = 5 * x + 15 * x**2 + 2 * x**3 + i*np.random.randn(number_of_samples, 1)
    y = y.flatten()
    z = np.polyfit(x1,y,m)
    p = np.poly1d(z)
    x2 = np.linspace(-15,15,50)
    plt.figure()
    plt.plot(x1, y,'ro',x2,p(x2),'b--')
    plt.title('Plot of the polynomial with noise_scale : %d' %i)
    plt.legend(['noisy data','polynomial'])
```

The increase in noise scale, increasingly distorts the plot of the polynomial. This can be evidently observed in the plots above. When this data is used to train polyfit(), the predicted polynomial is also impacted, which inturn affects the model accuracy. Thus increasing noise in the data also increasingly affects the model and decreases its accuracy. This might further lead to overfitting.
